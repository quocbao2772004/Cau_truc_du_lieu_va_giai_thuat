```python
#set the directory

dir = '/content/drive/MyDrive/AI/crawl image from web/data_crawl_web'
```

```python
downloader.download('tiger animal', limit = 75, output_dir = dir, adult_filter_off = False)
```

```python
import os
```

```python
image_folder='/content/drive/MyDrive/AI/crawl image from web/data_crawl_web/tiger animal'
image_file=os.listdir(image_folder)
```

```python
for i,file_name in enumerate(image_file):
  new_name=f'tiger_{i+1}.jpg'
  old_path=os.path.join(image_folder,file_name)
  new_path=os.path.join(image_folder,new_name)
  os.rename(old_path,new_path)
print("complete")
```
